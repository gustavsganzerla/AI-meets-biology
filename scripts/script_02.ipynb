```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
###import the data created by the first script

promoters_df = pd.read_csv('/home/user100/scratch/data/promoters.csv', header = None)
control_df = pd.read_csv('/home/user100/scratch/data/control.csv', header = None)

promoters_mean = promoters_df.mean()
control_mean = control_df.mean()
```

```python
###lets check our data

plt.figure(figsize=(12,8))

plt.plot(promoters_mean, label='Promoters', marker='o')
plt.plot(control_mean, label='Controls', marker='x')

plt.legend()

plt.show()
```

```python
### for a binary classification, we need to add the labels

promoters_df['label'] = 1
control_df['label'] = 0

```

```python
###finally, let's concatenate both dfs 
df = pd.concat([promoters_df,control_df], ignore_index=True)

df.to_csv('/home/user100/scratch/data/all.csv', index=False)
```
